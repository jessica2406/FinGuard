```python
import pandas as pd
import os
import shutil

# 1. Locate the file (from your kagglehub output)
raw_data_path = ".\data\PS_20174392719_1491204439457_log.csv" 
# Note: Replace the above with the actual path printed by kagglehub

# 2. Load the data
df = pd.read_csv(raw_data_path)

# 3. Focus on suspicious types: TRANSFER and CASH_OUT
# Also, let's keep ALL known fraud cases and a sample of non-fraud cases
fraud_cases = df[df['isFraud'] == 1]
non_fraud_sample = df[df['isFraud'] == 0].sample(n=15000, random_state=42)

# 4. Combine and Save
working_df = pd.concat([fraud_cases, non_fraud_sample])
os.makedirs('data', exist_ok=True)
working_df.to_csv('data/cleaned_transactions.csv', index=False)

print(f"Data ready! Total records: {len(working_df)}")
print(f"Fraudulent records included: {len(fraud_cases)}")
```
