```python
import pandas as pd
from neo4j import GraphDatabase
import os

# --- CONNECTION ---
uri = "neo4j+s://2829b1f2.databases.neo4j.io"
username = "2829b1f2"   # ✅ IMPORTANT: use instance ID (your working one)
password = os.getenv("NEO4J_PASSWORD") 

driver = GraphDatabase.driver(uri, auth=(username, password))

# ✅ Test connection
driver.verify_connectivity()
print("✅ Connected to Neo4j Aura")

# --- LOAD DATA ---
df = pd.read_csv('../data/cleaned_transactions.csv')  # ✅ safer path

print(f"📊 Total rows: {len(df)}")

# --- UPLOAD FUNCTION ---
def upload_transactions(tx, data):
    query = """
    UNWIND $rows AS row
    MERGE (sender:Client {id: row.nameOrig})
    MERGE (receiver:Client {id: row.nameDest})
    MERGE (sender)-[t:SENT_MONEY_TO]->(receiver)
    SET t.amount = coalesce(row.amount, 0),
        t.type = row.type,
        t.step = row.step,
        t.isFraud = coalesce(row.isFraud, 0)
    """
    tx.run(query, rows=data)

# --- BATCH UPLOAD ---
batch_size = 1000

with driver.session(database="2829b1f2") as session:   # ✅ explicit DB (safer)
    for i in range(0, len(df), batch_size):
        batch = df.iloc[i:i+batch_size].to_dict('records')
        session.execute_write(upload_transactions, batch)
        print(f"✅ Uploaded {i + len(batch)} rows...")

print("🚀 Phase 3 Complete: Knowledge Graph is live!")

driver.close()
```
