```python
import pandas as pd
from neo4j import GraphDatabase

# 1. Setup Connection Details
uri = "bolt://localhost:7687"
username = "neo4j"
password = "password123" # Use the password you just typed in Neo4j

driver = GraphDatabase.driver(uri, auth=(username, password))

# 2. Load our cleaned data
df = pd.read_csv('data/cleaned_transactions.csv')

def upload_transactions(tx, data):
    query = """
    UNWIND $rows AS row
    MERGE (sender:Client {id: row.nameOrig})
    MERGE (receiver:Client {id: row.nameDest})
    CREATE (sender)-[t:SENT_MONEY_TO]->(receiver)
    SET t.amount = row.amount,
        t.type = row.type,
        t.step = row.step,
        t.isFraud = row.isFraud
    """
    tx.run(query, rows=data)

# 3. Batch Upload
batch_size = 1000
with driver.session() as session:
    for i in range(0, len(df), batch_size):
        batch = df.iloc[i:i+batch_size].to_dict('records')
        session.execute_write(upload_transactions, batch)
        print(f"Uploaded {i + len(batch)} rows...")

print("Phase 3 Complete: Knowledge Graph is live!")
driver.close()
```
