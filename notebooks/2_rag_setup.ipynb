```python
import os
from langchain_community.document_loaders import PyPDFLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
from langchain_huggingface import HuggingFaceEmbeddings
from langchain_community.vectorstores import Chroma

# 1. Load the PDF
pdf_path = "./docs/aml_rules.pdf"
loader = PyPDFLoader(pdf_path)
documents = loader.load()

# 2. Split the text into chunks
# We use chunks because LLMs have a limit on how much text they can read at once.
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000, 
    chunk_overlap=150
)
chunks = text_splitter.split_documents(documents)

# 3. Choose an Embedding Model
# This turns text into numbers (vectors) that the AI understands.
# 'all-MiniLM-L6-v2' is fast and runs perfectly on a local laptop.
embeddings = HuggingFaceEmbeddings(model_name="all-MiniLM-L6-v2")

# 4. Create and Save the Vector Database
# This will create a folder named 'vector_db' in your project directory
persist_directory = "vector_db"

vector_db = Chroma.from_documents(
    documents=chunks,
    embedding=embeddings,
    persist_directory=persist_directory
)

print(f"Phase 2 Complete! Vector DB created with {len(chunks)} text chunks.")
```

```python
query = "What are the red flags for suspicious transactions?"
docs = vector_db.similarity_search(query, k=2)

for i, doc in enumerate(docs):
    print(f"\n--- Result {i+1} ---")
    print(doc.page_content[:500] + "...")
```
